```python
# Imports
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Load the data
from google.colab import files
uploaded = files.upload()
```

```python
# Store the data
NVDA = pd.read_csv('NVDA.csv')
# Show the data
NVDA
```

```python
# Visualize the data
plt.figure(figsize=(12.5,4.5))
plt.plot(NVDA['Adj Close'], label='NVDA')
plt.title('NVDA Adj. Close Price History')
plt.xlabel('Oct - Dec')
plt.ylabel('Adj. Close Price $')
plt.legend(loc='upper left')
plt.show()
```

```python
# Create the simple moving average with a 30 day window
SMA30 = pd.DataFrame()
SMA30['Adj Close'] = NVDA['Adj Close'].rolling(window=30).mean()
SMA30
```

```python
# Create a SMA 100 days
SMA100 = pd.DataFrame()
SMA100['Adj Close'] = NVDA['Adj Close'].rolling(window=100).mean()
SMA100
```

```python
# Vizualize the data
plt.figure(figsize=(12.5,4.5))
plt.plot(NVDA['Adj Close'], label='NVDA')
plt.plot(SMA30['Adj Close'], label='SMA30')
plt.plot(SMA100['Adj Close'], label='SMA100')
plt.title('NVDA Adj. Close Price History')
plt.xlabel('Oct - Dec')
plt.ylabel('Adj. Close Price $')
plt.legend(loc='upper left')
plt.show()
```
